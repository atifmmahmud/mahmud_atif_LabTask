```python
import pandas as pd 
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import altair as alt

sns.set_theme(style="whitegrid", palette="pastel")
```

```python
df_music_streams = pd.read_csv("music_streams.csv")
display(df_music_streams)
```

```python
display(df_music_streams.columns)
```

# Histogram

```python
# Create a histogram
sns.histplot(data=df_music_streams, x="streams_millions", bins=20)
```

```python
# Look at min max etc
df_music_streams.describe()
```

```python
median = df_music_streams.streams_millions.median()
mean = df_music_streams.streams_millions.mean()

# Create a histogram with ouitlier removed
fig, ax = plt.subplots(figsize=(10,6))
sns.histplot(data=df_music_streams[df_music_streams.streams_millions < 100], x="streams_millions", bins=20, color="pink")
ax.axvline(median, color="darkorange", label="median")
ax.axvline(mean, color="darkblue", label="mean")
ax.set_title("Monthly Streams")
ax.set_xlabel("Monthly streams (millions)")
ax.set_ylabel("Song Count")
ax.legend()
plt.show()
```

## Task from Prof. Karduni

```python
# Function to create a histogram:
def create_histogram(data_source, column, bins, title, xlabel, ylabel, color):
    fig, ax = plt.subplots(figsize=(12,6))
    sns.histplot(data=data_source, x=column, bins=bins, color=color)
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    plt.show()

create_histogram(df_music_streams, "chart_peak", 10, "Chart Peaks", "Chart Peaks", "Count", "darkblue")
```

# Boxplots

```python
order = df_music_streams.groupby("genre")["energy"].median().sort_values().index
print(order)
sns.boxplot(data=df_music_streams, y="energy", x="genre", order=order)
```

```python
# Add a log monthly listeners column
df_music_streams["monthly_listeners_log"] = np.log(df_music_streams["monthly_listeners"])
```

```python
plt.figure(figsize=(20, 10))
sns.boxplot(data=df_music_streams, y="monthly_listeners", x="genre")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
sns.boxplot(data=df_music_streams, y="monthly_listeners", x="genre")
ax.set_yscale("log", base=10)
plt.show()
```

# Final Task: Altair viz
### Visualize association between genre and speechiness

```python
chart = alt.Chart(df_music_streams).mark_circle(size=20).encode(
    alt.X("genre").title("Genre"), # X-axis is Genre
    alt.Y("speechiness").title("Speechiness").scale(domain=(0, max(df_music_streams["speechiness"]))), # Y axis speechiness, make it scale to max to look nice
    tooltip=["speechiness", "genre", "streams_millions", "title"] # Data to show on hover of point
).properties(
    title="Speechiness of tracks by genre",
    width = 800,
    height = 480
).interactive() # Make it interactive

chart.display()
```
