### Part 3: Filtering ###

1. Show only the rows where the age is above 60.

2. Show rows where the fare is above 100 and the passenger was in 1st class.

3. Identify and isolate rows with missing values in the age column.

```python
# Show only rows where age is above 60
senior_citizens = df.loc[df["Age"] > 60]
print("3.1 Passengers above age 60")
display(senior_citizens)

# Space
print("\n\n")

# Show only roes where fare is above 100 and passenger was in 1st class
rich_people = df.loc[(df["Fare"] > 100) & (df["Pclass"] == 1)]
print("3.2 Passengers with fare above 100 and was in 1st class")
display(rich_people)

# Space
print("\n\n")

# Isolate rows with missing values in age column
missing_age = df.loc[df["Age"].isnull()]
print("3.3 Rows with missing age given below")
display(missing_age)

# Space
print("\n")

print("I can print out the remaining rows by removing such rows where age is missing")
filtered_table = df[df["Age"].notnull()]
display(filtered_table)

```

### Part 5: Arithmetic and Data Alignment ###

1. Create a new column that converts fares from USD to Euros.

```python
# 1 USD = 0.86 Euro according to Google
print("5.1 Creating a new column showing fares in Euros. 1 USD = 0.86 Euros")
df["Fare in Euros"] = df["Fare"] * 0.86
display(df)

# Space
print("\n\n")

# Subtract age
print("5.2 Table with average age subtracted from every age value. Let's save the filtered df to a backup for future calculations.")
df_2 = filtered_table.copy()
avg_age = df["Age"].mean()
df["Age"] = df["Age"] - avg_age
display(df)

# Space
print("\n\n")

# Subtract a custom array from two selected columns
# This is an array with value[i] = i * 2
print("5.3 Subtracting custom array from 2 selected columns (PassengerId and Fare). Array is such that value[i] = i * 2")
custom_array_1 = [x * 2 for x in range(891)]

df["PassengerId"] = df["PassengerId"] - custom_array_1
df["Fare"] = df["Fare"] - custom_array_1
display(df)


```

1. Create a new column with the number of characters in each name.
