### Part 1: Data Exploration ###
1. Print the shape and column names of the dataset.

2. Display basic information and summary statistics about the dataset.

3. Show the first 10 rows.

```python
# Formatting the string as per: https://www.w3schools.com/python/ref_string_format.asp
# Converting the columns pandax index to a basic list so it can easily be printed; otherwise it gave "dtype = str" at the end. Used the help of Claude for this.
print("1.1  The shape of the dataset is {shape} and the columns are {columns}".format(shape = df.shape, columns = list(df.columns)))

# Space
print("\n\n")

# Summary statistics. Data type, description, and where applicable, min, max, mean, and std. dev.
# Got some information from here: https://www.kaggle.com/datasets/yasserh/titanic-dataset
df_summary = pd.DataFrame(index = df.columns, columns = ['Data Type', 'Description', 'Min', 'Max', 'Mean', 'Standard Deviation'])

# Construct a summary data frame from the original data frame
df_summary.loc["PassengerId"] = ["Categorical", "ID of the passenger", "N/A", "N/A", "N/A", "N/A"]
df_summary.loc["Survived"] = ["Categorical", "Whether or not the passenger survived. 0 = no. 1 = yes", "N/A", "N/A", "N/A", "N/A"]
df_summary.loc["Pclass"] = ["Categorical", "Which class the passenger was in: 1nd, 2nd, or 3rd", "N/A", "N/A", "N/A", "N/A"]
df_summary.loc["Name"] = ["Categorical", "Name of the passenger", "N/A", "N/A", "N/A", "N/A"]
df_summary.loc["Sex"] = ["Categorical", "Sex of the passenger. male or female.", "N/A", "N/A", "N/A", "N/A"]
df_summary.loc["Age"] = ["Numerical", "Age of the passenger", df["Age"].min(), df["Age"].max(), df["Age"].mean(), df["Age"].std()]
df_summary.loc["SibSp"] = ["Numerical", "Number of siblings/spouses aboard the titanic", df["SibSp"].min(), df["SibSp"].max(), df["SibSp"].mean(), df["SibSp"].std()]
df_summary.loc["Parch"] = ["Numerical", "Number of parents/children aboard the titanic", df["Parch"].min(), df["Parch"].max(), df["Parch"].mean(), df["Parch"].std()]
df_summary.loc["Ticket"] = ["Categorical", "Ticket Number", "N/A", "N/A", "N/A", "N/A"]
df_summary.loc["Fare"] = ["Numerical", "Fare paid", df["Fare"].min(), df["Fare"].max(), df["Fare"].mean(), df["Fare"].std()]
df_summary.loc["Cabin"] = ["Categorical", "Cabin Number", "N/A", "N/A", "N/A", "N/A"]
df_summary.loc["Embarked"] = ["Categorical", "Port of Embarkation: C = Cherbourg, Q = Queenstown, S = Southampton.", "N/A", "N/A", "N/A", "N/A"]

# Print it
print("1.2 The summary and descriptive data is given below.")
display(df_summary)


# Space
print("\n\n")

# The first 10 rows of the dataframe
print("1.3 The first 10 rows are given below")
display(df.head(10))
```
